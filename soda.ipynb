```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split

import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
datos = pd.read_csv("D:/3.-InteligenciaClase/Dataset/soda.csv",delimiter=";")
#? datos = pd.read_csv("../3.-InteligenciaClase/Dataset/soda.csv",delimiter=';')
datos
```

```python
d = {"si": 1, "no":0}
datos["GustaPopcorn"] =  datos["GustaPopcorn"].map(d)
datos["GustaSoda"] =  datos["GustaSoda"].map(d)
datos["GustaCine"] =  datos["GustaCine"].map(d)
datos
```

```python
vd = datos[["GustaPopcorn","GustaSoda","Edad"]].values
vi = datos["GustaCine"]
```

```python
clasificador = tree.DecisionTreeClassifier()
clasificador.fit(vd,vi)
prediccion = clasificador.predict([[1,1,7]])
if(prediccion == 0):
    print("No le Gusta")
else: print("Le Gusta")
```

```python
clase = ["No le Gusta", "Le Gusta"]
vi = ["GustaPopcorn","GustaSoda","Edad"]
```

```python
datos = tree.export_graphviz(clasificador,feature_names=vi,class_names=clase,filled=True,impurity=False)
graht = pydotplus.graph_from_dot_data(datos)
graht.write_png("D:/3.-InteligenciaClase/imagenes/sodaaaaa.png")
imagen = pltimg.imread("D:/3.-InteligenciaClase/imagenes/sodaaaaa.png")
plt.imshow(imagen);
```
