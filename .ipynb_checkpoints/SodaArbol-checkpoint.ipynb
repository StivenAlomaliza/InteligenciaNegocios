```python
import pandas  as pd 
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
datos = pd.read_csv("soda.csv",delimiter=";")
datos
```

```python
d = {"si":1,"no":0}
datos["GustaPopcorn"] = datos["GustaPopcorn"].map(d)
datos["GustaSoda"] = datos["GustaSoda"].map(d)
datos["GustaCine"] = datos["GustaCine"].map(d)
datos
```

```python
vi = datos[["GustaPopcorn","GustaSoda","Edad"]].values
vd = datos["GustaCine"]
```

```python
clasificador = tree.DecisionTreeClassifier()
clasificador.fit(vi,vd)
```

```python
prediccion = clasificador.predict([[0,1,18]])
if(prediccion == 0):
    print("No Le Gusta")
else: print("Le Gusta")
```

```python
import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
clase = ["No Le Gusta","Le Gusta"]
vi = ["GustaPopcorn","GustaSoda","Edad"]
```

```python
datos = tree.export_graphviz(clasificador,feature_names=vi,class_names=clase,filled=True,impurity=False)
graht = pydotplus.graph_from_dot_data(datos)
graht.write_png("practica.png")
imagen = pltimg.imread("practica.png")
plt.imshow(imagen);
```
