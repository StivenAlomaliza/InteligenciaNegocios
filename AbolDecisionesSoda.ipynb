```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
datos = pd.read_csv("D:/3.-InteligenciaClase/Dataset/soda.csv",delimiter=";")
datos
```

```python
d = {"si":1,"no":0}
datos["GustaPopcorn"] = datos["GustaPopcorn"].map(d) 
datos["GustaSoda"] = datos["GustaSoda"].map(d)
datos["GustaCine"] = datos["GustaCine"].map(d) 
datos
```

```python
vi = datos[["GustaPopcorn","GustaSoda","Edad"]]
vd = datos["GustaCine"]
```

```python
clasificador = tree.DecisionTreeClassifier()
```

```python
#Entrenar al algoritmo de clasificasion
clasificador.fit(vi, vd)
```

```python
# #?from sklearn import tree
import pydotplus
import matplotlib.pyplot as plt 
import matplotlib.image as pltimg 

clase = ["NoGustaCine","GustaCine"]
vi = ["GustaPopcorn","GustaSoda","Edad"]

datos = tree.export_graphviz(clasificador, feature_names= vi, class_names = clase, filled= True, impurity = False) ##! Cambia aspectos como el color o quita el Gini  
#datos = tree.export_graphviz(clasificador)
graph = pydotplus.graph_from_dot_data(datos)
graph.write_png("miarbolSODA.png")
imagen = pltimg.imread("miarbolSODA.png")
plt.imshow(imagen);
```
